```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_excel('Argentina.xlsx')
df
```

```python
df.info()
```

```python
#FILTRO POR OBJETO
host_name=df[(df ["host_name"] == "Elbio Mariano") | (df ["host_name"] == "Marian") | (df ["host_name"] == "Cecilia") | (df ["host_name"] == "Federico") | (df ["host_name"] == "Martin") ]
host_name
```

```python
#convertir archivo filtrado a CSV
host_name.to_csv("host_name.csv")
```

```python
host_since = df[ (df["host_since"] <"2019-01-01")]
host_since
```

```python
#convertir archivo filtrado a CSV
host_since.to_csv("host_since.csv")
```

```python
tiempo_respuesta = df[ (df["host_response_time"] =="within a day")]
tiempo_respuesta
```

```python
#convertir archivo filtrado a CSV
tiempo_respuesta.to_csv("tiempo_respuesta.csv")
```

```python
#FILTRO POR COMPARACION
accommodates=df[df["accommodates"] > 4]
accommodates
```

```python
#convertir archivo filtrado a CSV
accommodates.to_csv("accommodates.csv")
```

```python
#FILTRO POR OBJETO
room_type=df[(df ["room_type"] == "Entire home/apt")  | (df ["room_type"] == "Private room") ]
room_type
```

```python
#convertir archivo filtrado a CSV
room_type.to_csv("tipo_cuarto.csv")
```

```python
#FILTRO POR COMPARACION
beds=df[df["beds"] < 5]
beds
```

```python
#convertir archivo filtrado a CSV
beds.to_csv("camas.csv")
```
